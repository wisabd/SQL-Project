```python
## This file is to map the national population density of Pakistan to that of Punjab via Ray Tracing Algorithm, applied on the boundary coordinates of the Punjab region
```

```python
import pandas as pd
df = pd.read_csv('/Users/muhammadwisalabdullah/Downloads/pak_pd_2020_1km_ASCII_XYZ.csv')
# Rename specific columns (using rename)
df = df.rename(columns={"X": "Longitude", "Y": "Latitude", "Z": "Population Density at 1km"})
print(df)
```

```python
high_density_areas = df[df['Population Density at 1km'] > 350]
# To select specific columns (e.g., X, Y, Z) for these rows:
result = df.loc[df['Population Density at 1km'] > 300, ['Longitude', 'Latitude', 'Population Density at 1km']]
latitudes = df['Latitude'].tolist()
longitudes = df['Longitude'].tolist()
# Display results
print(result)
print(len(result))
```

```python
import requests
# Define the API URL
url = "https://nominatim.openstreetmap.org/search.php?q=Punjab,+Pakistan&polygon_geojson=1&format=json"
import urllib.request, json
with urllib.request.urlopen("https://nominatim.openstreetmap.org/search.php?q=Punjab,+Pakistan&polygon_geojson=1&format=json") as url:
    data = json.load(url)
    edges = data[0]['geojson']['coordinates']
```

```python
for i in range(len(edges[0][0])):
    print(edges[0][0][i])
```

```python
Constructed_Edges = []
for j in range(len(edges[0][0]) -1):
    Constructed_Edges.append([edges[0][0][j], edges[0][0][j+1]])
   # (ds, ds2), (dy1, dy2) = edges[0][0][j], edges[0][0][j + 1]
print(Constructed_Edges[5])
```

```python
def IsPunjab(edgess, xp, yp):
    cnt = 0
    for cv in edgess:
        (y1, x1), (y2, x2) = cv
        if (yp < y1) != (yp < y2) and xp < x1 + ((yp - y1) * (x2 - x1)):
             cnt = cnt + 1
    return cnt%2 == 1
```

```python
Truess = 0
Falses = 0
Punjab_coordinates = []
for i in range(len(latitudes)):
   if IsPunjab(Constructed_Edges, latitudes[i], longitudes[i]):
        Punjab_coordinates.append([latitudes[i], longitudes[i]])
```

```python
Punjab_lats = [pair[0] for pair in Punjab_coordinates]
Punjab_longs = [pair[1] for pair in Punjab_coordinates]
Punjab_df = df[
    df['Latitude'].isin(Punjab_lats) &
    df['Longitude'].isin(Punjab_longs)
]
```
